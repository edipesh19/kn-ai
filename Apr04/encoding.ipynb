```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder

# Create dummy data with categorical features
data = {
    'color': ['red', 'blue', 'green', 'red', 'blue'],
    'size': ['small', 'large', 'medium', 'large', 'small'],
    'price': [10, 20, 15, 25, 12]
}

df = pd.DataFrame(data)
print("Original Data:")
print(df)
print("\n")

# Method 1: Using pandas get_dummies
print("Method 1: Using pandas get_dummies")
df_encoded = pd.get_dummies(df, columns=['color', 'size'], drop_first=False)
print(df_encoded)
print("\n")

# Method 2: Using sklearn OneHotEncoder
print("Method 2: Using sklearn OneHotEncoder")
encoder = OneHotEncoder(sparse_output=False)
encoded_array = encoder.fit_transform(df[['color', 'size']])
encoded_df = pd.DataFrame(encoded_array, columns=encoder.get_feature_names_out(['color', 'size']))
print(encoded_df)
```
